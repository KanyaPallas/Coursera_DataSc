```python
from numpy import sin, exp
def f(x):
    x =float(x)
    return sin(x/5)*exp(x/10)+5*exp(-x/2)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace (1, 15)
y = [f(t) for t in x] 
plt.plot(x, y)
```

```python
#Первая степень
ma = np.array([[1, 1], [1, 15]])
mb = np.array([f(1), f(15)])
w = np.linalg.solve(ma, mb)
print w
```

```python
x = np.linspace(1, 15)
y = [w[0]+w[1]*t for t in x]
plt.plot(x, y)
z = [f(t) for t in x] 
plt.plot(x, z)
```

```python
#Вторая степень
x1 = 1
x2 = 8
x3 = 15
ma2 = ([1, x1, x1**2],[1, x2, x2**2],[1, x3, x3**2])
mb2 = ([f(x1), f(x2), f(x3)])
w2 = np.linalg.solve(ma2, mb2)
print w2
```

```python
x = np.linspace(1, 15)
y = [w2[0]+w2[1]*t+w2[2]*t**2 for t in x]
plt.plot(x, y)
z = [f(t) for t in x] 
plt.plot(x, z)
```

```python
x1 = 1.0
x2 = 4.0
x3 = 10.0
x4 = 15.0
ma3 = ([1, x1, x1**2, x1**3],[1, x2, x2**2, x2**3],[1, x3, x3**2, x3**3], [1, x4, x4**2, x4**3])
mb3 = ([f(x1), f(x2), f(x3), f(x4)])
w3 = np.linalg.solve(ma3, mb3)
print w3.round(2)
```

```python
x = np.linspace(1, 15)
y = [w3[0]+w3[1]*t+w3[2]*t**2+w3[3]*t**3  for t in x]
plt.plot(x, y)
z = [f(t) for t in x] 
plt.plot(x, z)
```
